```python
from aocd import get_data
import numpy as np

day17 = get_data(day=17,year=2021)
```

```python
class probe:
    x = 0
    y = 0
    xvel = 0
    yvel = 0
    xmin = 0
    xmax = 0
    ymin = 0
    ymax = 0
    
    def step(self):
        self.x += self.xvel
        self.y += self.yvel
        self.xvel -= np.sign(self.xvel)
        self.yvel -= 1
        
    def hitstarget(self,xvel,yvel):
        self.x = 0
        self.y = 0
        self.xvel = xvel
        self.yvel = yvel
        nsteps = 0
        maxheight = 0
        while self.y > self.ymin:
            nsteps += 1
            self.step()
            if self.y > maxheight: maxheight = self.y
            if (self.xmin <= self.x <= self.xmax) & (self.ymin <= self.y <= self.ymax):
                return True, nsteps, maxheight
        return False,0,0
        
```

```python
p = probe()
p.xmin = 20
p.xmax = 30
p.ymin = -10
p.ymax = -5
p.hitstarget(6,9)
```

```python
p = probe()
target = day17.replace('=',' ').replace(',',' ').replace('.',' ').split(' ')
p.xmin = int(target[3])
p.xmax = int(target[5])
p.ymin = int(target[8])
p.ymax = int(target[10])
```

```python
p.hitstarget(10,10)
```

```python
results = []
maxheight = 0
for i in range(100):
    for j in range(-300,300):
        r = p.hitstarget(i,j)
        if r[0]: results.append([i,j,r])
        if r[2] > maxheight: maxheight = r[2]
print(maxheight)
```
